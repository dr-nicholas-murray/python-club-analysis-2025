```python
# Mount Google Drive and import pandas
from google.colab import drive
drive.mount('/content/gdrive')
import pandas as pd
```

```python
# Define the file path and load the dataset into a DataFrame named 'df'
file_path = ('/content/gdrive/MyDrive/VCCRI_Python_Club_2025/'
            'data/pneumonia_recovery_dataset.csv')
df = pd.read_csv(file_path)
print ("Setup complete. Main DataFrame 'df' is ready for analysis.")
df.head()
```

```python
# Use the .rename() method to change column names
df.rename(columns={'patient_id':'subject_id'}, inplace=True)
print("Column 'patient_id ' has been renamed:")
df.head()
```

```python
print("---Descriptive Stats WITH Duplicates---")
df.describe()
```

```python
# First , verify the number of duplicates
num_duplicates = df.duplicated().sum()
print (f"Number of duplicate rows found: {num_duplicates}")
```

```python
# Now , remove them . 'inplace=True' modifies 'df' directly.
df.drop_duplicates(inplace=True)
print(f"Shape after dropping duplicates: {df. shape }")
```

```python
print("\n---Descriptive Stats AFTER Removing Duplicates---")
df.describe()
```

```python
# Step A: Calculate the median of the 'bmi' column
median_bmi = df['bmi'].median()
print(f"The median BMI to be used for filling is: {median_bmi}")
df.head(20)
```

```python
# Step B: Use the .fillna() method to fill the missing spots (permanetly)
df["bmi"] = df["bmi"].fillna(median_bmi)
df.head(20)
```

```python
# Step C: Verify that the missing values are gone
print("\nMissing values count after filling:")
print(df.isnull().sum())
```

```python
# --- Simulation Start ---
print("The original data type of 'age':")
print(df.dtypes)
```

```python
# Deliberately change the 'age' column to a string type
df['age'] = df['age'].astype(str)
print(" Data type of 'age' is now object (string):")
print(df.dtypes)
```

```python
# --- Simulation End ---
# Now , let 's convert it back to a numeric ( integer ) type
# pd. to_numeric () is a safe way to do this
df['age'] = pd.to_numeric(df['age'])
print("\nData type of 'age' is now correctly integer:")
print (df.dtypes)
```

```python
# Get the index label of the first row in our DataFrame
first_row_index = df.index[0]
df.head(3)
```

```python
# Use .loc to target the cell by its row and column labels
df.loc[first_row_index, 'sex'] = 'male'
df.head(3)
```
